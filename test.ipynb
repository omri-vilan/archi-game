```python
import zipfile
import xml.etree.ElementTree as ET
import re

xlsx_path = "data.xlsx"

with zipfile.ZipFile(xlsx_path) as z:

    # Parse image relationships
    rels_xml = z.read(
        "xl/richData/_rels/richValueRel.xml.rels"
    )

    rels_root = ET.fromstring(rels_xml)

    rid_to_image = {}

    for rel in rels_root:
        rid = rel.attrib["Id"]
        target = rel.attrib["Target"]

        image_name = target.split("/")[-1]
        rid_to_image[rid] = image_name

    # Parse richValueRel.xml
    rv_xml = z.read("xl/richData/richValueRel.xml")
    rv_root = ET.fromstring(rv_xml)

    vm_to_image = {}

    for idx, rel in enumerate(rv_root):
        rid = rel.attrib[
            "{http://schemas.openxmlformats.org/officeDocument/2006/relationships}id"
        ]

        vm = idx + 1
        vm_to_image[vm] = rid_to_image[rid]

    # Parse worksheet
    sheet_xml = z.read("xl/worksheets/sheet1.xml").decode()

    matches = re.findall(
        r'<c[^>]*r="([^"]+)"[^>]*vm="(\d+)"',
        sheet_xml
    )

    for cell, vm in matches:
        vm = int(vm)

        print(
            f"{cell} -> {vm_to_image.get(vm)}"
        )
```
